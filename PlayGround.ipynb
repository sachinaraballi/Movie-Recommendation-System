```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("titles.csv")
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

Mean Calculation

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['number of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['number of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='number of ratings',data=ratings,alpha=0.5)
```

People watched good movies more and rated them good we can see movies with 0 to 2 reviewed by less number of viewers

```python
movie_matrix = df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
movie_matrix
```

```python
ratings.sort_values('number of ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
starwars_user_ratings = movie_matrix['Star Wars (1977)']
liarliar_user_ratings = movie_matrix['Liar Liar (1997)'] # we chose stars wars and liar liar movie
```

```python
starwars_user_ratings.head()
```

```python
similar_to_starwars = movie_matrix.corrwith(starwars_user_ratings)
similar_to_liarliar = movie_matrix.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head() #removing nan values 
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['number of ratings'])
corr_starwars.head()
```

Let's fix this by filtering out movies that have less than 100 reviews (this value was chosen based off the histogram from earlier).

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['number of ratings'])
corr_liarliar[corr_liarliar['number of ratings']>100].sort_values('Correlation',ascending=False).head()
```
